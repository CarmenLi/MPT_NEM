```python
for bs in range(100):

    plt.plot(EF[bs].iloc[-2], EF[bs].iloc[-1])
plt.show()
```

```python
# find common points

# initialise
v_com=EF[0].loc['cost'].values
for i in np.arange(1,100):
    v_com=np.intersect1d(v_com, EF[i].loc['cost'].values)
v_com
```

```python
M_samples=np.hstack([
    np.array([EF[bs][1][33] for bs in range(100)]).reshape(-1,1), 
    np.vstack([ (EF[bs].loc['sigma_sq'][EF[bs].loc['cost'].isin(v_com)].values) for bs in range(100)]),
    np.array([EF[bs].iloc[34, -1] for bs in range(100)]).reshape(-1,1)
])
```

```python
M_samples.shape
```

```python
from scipy import stats
# compute absolute of zscore along each column
zM_samples=stats.zscore(M_samples, axis=0)
azM_samples=np.abs(zM_samples)
```

```python
plt.plot(azM_samples.T)
plt.show()
```

```python
s_mean_azscore=pd.Series(index=range(100), data=azM_samples.mean(axis=1))
s_mean_azscore.nsmallest(10)
```

```python
s_mean_azscore
```

```python
s_mean_azscore=pd.Series(index=range(100), data=azM_samples[:, :-1].mean(axis=1))
s_mean_azscore.nsmallest(10)
```

```python
s_mean_azscore.plot()
```
